```python
import pandas as pd
import yfinance as yf

stock_symbol = 'KO'
start_date = '2022-01-01'
end_date = '2022-07-30'

coca_cola = yf.download(stock_symbol, start_date, end_date)
```

```python
coca_cola.tail()
```

#### adjusted closing price
instead of closing price (cash value of the last transacted price before the market closes), adjusted closing price factors in *corporate actions such as splits, dividends, and rights offerings*

```python
coca_cola = yf.download('KO', '2022-03-19', '2022-07-30', auto_adjust=True)
coca_cola.tail()
```

```python
coca_cola.index = pd.to_datetime(coca_cola.index)
coca_cola.tail()
```
